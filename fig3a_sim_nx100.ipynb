```python
from matplotlib.mlab import griddata

Ngrid = 1000
xi = np.linspace(6, 9, Ngrid)
yi = np.linspace(1, 4, Ngrid)

x = np.log10(np.array([r.parameters['kMinus'] for r in annular53Records]))
y = np.log10(np.array([r.parameters['kPlus'] for r in annular53Records]))
z = np.array([r.void_size for r in annular53Records])

#zi = griddata((x, y), z, (xi[None,:], yi[:,None]), method='cubic')
zi_void = griddata(x, y, z[:,0], xi, yi)
zi_height = griddata(x, y, z[:,1], xi, yi)

fig = plt.figure(figsize=(10, 6))

ax1 = fig.add_subplot(1, 3, 1)

ax1.contourf(10**xi, 10**yi, zi_void, np.linspace(0, 1, 11))

ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlabel(r'$k^-')
ax1.set_ylabel(r'$k^+')
#plt.colorbar()
plt.title(r'Void Size')


ax2 = fig.add_subplot(1, 3, 2)

ax2.contourf(10**xi, 10**yi, zi_height, np.linspace(0, 1, 11))

ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlabel(r'$k^-')
ax2.set_ylabel(r'$k^+')
plt.title(r'Fill Height')

ax3 = fig.add_subplot(1, 3, 3)

ax3.contourf(10**xi, 10**yi, zi_height - zi_void, np.linspace(0, 1, 11))

ax3.set_xscale('log')
ax3.set_yscale('log')
ax3.set_xlabel(r'$k^-')
ax3.set_ylabel(r'$k^+')
plt.title(r'(Fill Height) - (void size)')
```
