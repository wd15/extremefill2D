# Contour Plot

```python
%load_ext autoreload
%autoreload 2
from tools import getSMTRecords, smt_ipy_table
records = getSMTRecords()
```

```python
annular53Records = getSMTRecords(records, tags=['annular53'])
```

```python
%load_ext autoreload
%autoreload 4
from tools import VoidSizeFinder

vv = VoidSizeFinder()
for r in annular53Records:
    r.void_size = vv.find_void_size(r)
```

```python
from matplotlib.mlab import griddata

titlefontsize = 18
fontdict = {'fontsize' : 16}
tickfontsize = 14
Ngrid = 1000
xi = np.linspace(6, 9, Ngrid)
yi = np.linspace(0, 4, Ngrid)

x = np.log10(np.array([r.parameters['kMinus'] for r in annular53Records]))
y = np.log10(np.array([r.parameters['kPlus'] for r in annular53Records]))
z = np.array([r.void_size for r in annular53Records])

#zi = griddata((x, y), z, (xi[None,:], yi[:,None]), method='cubic')
zi_void = griddata(x, y, z[:,0], xi, yi)
zi_height = griddata(x, y, z[:,1], xi, yi)

fig = plt.figure(figsize=(15, 6))


ax1 = fig.add_subplot(1, 3, 1)
ax1.contourf(10**xi, 10**yi, zi_void, np.linspace(0, 1, 11))
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlabel(r'$k^-', fontdict=fontdict)#, fontsize=labelfontsize)
ax1.set_ylabel(r'$k^+', fontdict=fontdict)#, fontsize=labelfontsize)
#ax1.set_yticklabels(fontdict=fontdictticks)
plt.title(r'(a)', fontsize=titlefontsize)
ax1.tick_params(axis='both', which='major', labelsize=tickfontsize)
ax1.tick_params(axis='both', which='minor', labelsize=tickfontsize)


ax2 = fig.add_subplot(1, 3, 2)
ax2.contourf(10**xi, 10**yi, zi_height, np.linspace(0, 1, 11))
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlabel(r'$k^-', fontdict=fontdict)
#ax2.set_ylabel(r'$k^+')
ax2.set_yticklabels([r''] * len(ax2.get_yticks()))
plt.title(r'(b)', fontsize=titlefontsize)
ax2.tick_params(axis='both', which='major', labelsize=tickfontsize)
ax2.tick_params(axis='both', which='minor', labelsize=tickfontsize)


ax3 = fig.add_subplot(1, 3, 3)
a = ax3.contourf(10**xi, 10**yi, zi_height - zi_void, np.linspace(0, 1, 11))
#plt.colorbar(a)
ax3.set_xscale('log')
ax3.set_yscale('log')
ax3.set_xlabel(r'$k^-', fontdict=fontdict)
#ax3.set_ylabel(r'$k^+')
ax3.set_yticklabels([r''] * len(ax3.get_yticks()))
plt.title(r'(c)', fontsize=titlefontsize)
ax3.tick_params(axis='both', which='major', labelsize=tickfontsize)
ax3.tick_params(axis='both', which='minor', labelsize=tickfontsize)


## colorbar
from mpl_toolkits.axes_grid1 import make_axes_locatable
divider = make_axes_locatable(ax3)
cax = divider.append_axes("right", size="5%", pad=0.05)
cb = plt.colorbar(a, cax=cax)
cb.ax.tick_params(labelsize=14) 

#plt.tight_layout()
plt.savefig("contour.png")
```

# Deposition Rate Versus Time

```python
%load_ext autoreload
%autoreload 2
from tools import getSMTRecords, smt_ipy_table
records = getSMTRecords()
```

```python
annular53Records = getSMTRecords(records, tags=['annular53'])
```

```python
from tools import getRecord

kPlus = np.logspace(1, 4, 20)[[5, 10, 15, 19]][::-1][2]
kMinuses = np.logspace(6, 9, 20)[[5, 10, 15, 19]][[0, 1, 2]]
reorderedRecords = [getRecord(records=annular53Records, kPlus=kPlus, kMinus=kMinus) for kMinus in kMinuses]
```

```python
from tools import getDepositionRates

ticklabelfontsize = 14
axislabelfontsize = 16
legendfontsize = 14

labels = ['passivated', 'activated', 'voided']
linestyles = ['--', '-', ':']

for r, l, ls in zip(reorderedRecords, labels, linestyles):
    t, d = getDepositionRates(r)
    label = '{0} ($\\texttt{{{1}}}$)'.format(l, r.label[:8])
#    plt.semilogy(t, d, label=label, ls=ls, color='k', linewidth=2)
    a = plt.loglog(t, d, label=label, ls=ls, color='k', linewidth=2)
plt.ylim(1e-9, 1e-7)
plt.xlabel(r'$t$ $\left(\second\right)$', fontdict={'fontsize' : axislabelfontsize})
plt.ylabel(r'$v$ $\left(\meter\per\second\right)$', fontdict={'fontsize' : axislabelfontsize})
plt.xlim(6e1, 1e4)
plt.tick_params(axis='both', which='major', labelsize=tickfontsize)
plt.tick_params(axis='both', which='minor', labelsize=tickfontsize)
plt.legend(prop={'size' : legendfontsize})
```
