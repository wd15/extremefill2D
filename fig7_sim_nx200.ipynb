```python
%load_ext autoreload
%autoreload 2
from tools import getSMTRecords, smt_ipy_table
records = getSMTRecords(tags=['annular60'])
```

```python
print len(records)
```

```python
%load_ext autoreload
%autoreload 4
from featureProperty import FeatureProperty

suppressor = []
potential = []
void_size = []
height = []

for r in records:
    if len(r.output_data) > 0:
        fp = FeatureProperty(r)
        void_size.append(fp.getVoidSize())
        height.append(fp.getHeight())
        suppressor.append(r.parameters['bulkSuppressor'])
        potential.append(r.parameters['appliedPotential'])
        fp.close()
    
void_size = np.array(void_size)
height = np.array(height)
suppressor = np.array(suppressor)
potential = np.array(potential)

values = height - void_size
```

```python
from tools import refine_contour_plot
points = np.log10(np.array((suppressor, -potential)).transpose())
new_points = refine_contour_plot(points, values)[:100]
new_suppressor, new_potential = 10**new_points[:,0], 10**new_points[:,1]
```

```python
from scipy.spatial import Delaunay
ps = np.concatenate((points, new_points))
tri1 = Delaunay(points)
tri2 = Delaunay(ps)
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax1.triplot(points[:,0], points[:,1], tri1.vertices.copy())
ax1.plot(points[:,0], points[:,1], 'o', ms=3)
ax2 = fig.add_subplot(1, 2, 2)
ax2.triplot(ps[:,0], ps[:,1], tri2.vertices.copy())
ax2.plot(ps[:,0], ps[:,1], 'o', ms=3)
plt.show()
```

```python
!date

from tools import batch_launch

reason = "Fig 7, Nx=200, refinement 1"
    
for bulkSuppressor, appliedPotential in  zip(new_suppressor, -new_potential):
    batch_launch(reason=reason,
                 tags=['fig7_nx200_refinement1'],
                 kPlus=5e2,
                 kMinus=3e7,
                 rboundary=50e-6 / np.sqrt(np.pi),
                 Nx=200,
                 spacing_ratio=1.2,
                 CFL=0.1,
                 levelset_update_ncell=5,
                 data_frequency=100,
                 bulkSuppressor=bulkSuppressor,
                 appliedPotential=appliedPotential)
```

```python
from multiViewer import MultiViewer
from tools import getRecord
        
bulkSuppressors = 2 * np.logspace(-3, -1, 20)
appliedPotentials = -2.5 * np.logspace(-2, 0, 20)

reorderedRecords = [[getRecord(records=records, bulkSuppressor=bulkSuppressor, appliedPotential=appliedPotential) \
    for bulkSuppressor in bulkSuppressors] for appliedPotential in appliedPotentials]


def me(n):
    s = '%1.1e' % n
    m, e = s.split('e')
    return float(m), int(e)

def rowlabel(r):
    if r:
        m, e = me(r.parameters['appliedPotential'])
        return '${0:1.1f}\\times 10^{{{1}}}$'.format(m, e)
    else:
        return ''

def columnlabel(r):
    if r:
        m, e = me(r.parameters['bulkSuppressor'])
        return '${0:1.1f}\\times 10^{{{1}}}$'.format(m, e)
    else:
        return ''

viewer = MultiViewer(reorderedRecords,
                     rowtitle=rowlabel,
                     columntitle=columnlabel,
                     figsize=(1.5, 6))

viewer.plot(times=np.arange(20) * 5000. / 19.)
#
#smt_ipy_table([j for i in reorderedRecords for j in i if j],
#              fields=['label', 'timestamp', 'parameters', 'duration', 'version', 'tags'],
#              parameters=['bulkSuppressor', 'appliedPotential', 'kPlus', 'kMinus'])
```
