# $C_{\text{SUPP}}$ v $E_{\text{APPLIED}}$ contour plot with `Nx=200`

Simulations are to create a map of $C_{\text{SUPP}}$ v $E_{\text{APPLIED}}$ to figure out a better set of paramters that might give full fill inside the feature.

```python
from tools import batch_launch

!date
    
reason = "Running simulation for bulkSuppressor v appliedPotential contour plot with Nx=200"
    
for bulkSuppressor in (2 * np.logspace(-3, -1, 20)):
    for appliedPotential in (-2.5 * np.logspace(-2, 0, 20)):
        batch_launch(reason=reason,
                     tags=['annular59'],
                     kPlus=1e3,
                     kMinus=4e7,
                     rboundary=50e-6 / np.sqrt(np.pi),
                     Nx=200,
                     spacing_ratio=1.2,
                     CFL=0.1,
                     levelset_update_ncell=5,
                     data_frequency=100,
                     bulkSuppressor=bulkSuppressor,
                     appliedPotential=appliedPotential)
```
