```python
import pylab
pylab.figure()

for index in range(4):
    CFLs = []
    norms = []
    for t, v, c in data:
        CFLs.append(c.strip('CFL'))
        norms.append(v[index])
     
    pylab.loglog(CFLs, norms, label='$t$=%1.1e' % t[index])

pylab.xlabel('CFL')
pylab.ylabel(r'$||\phi - \phi_0||$')
pylab.loglog(CFLs, CFLs, 'k--', label='$O(1)$')
pylab.legend()
pylab.show()
```

Below values of CFL = 0.1 the convergence seems to be first order. This doesn't seem to be true for the red curve, however. There could be issues with the number of cells being used in the Norm evaluation. Currently the Norm evaluation is only for cells within a distance of 10 cell widths from the zero level set. I think the evidence of convergence is good enough to proceed onto looking at grid convergence. A CFL less than 0.1 may be needed to provide sufficient accuracy.
