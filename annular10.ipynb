# $C_{\text{SUPP}}$ v $E_{\text{APPLIED}}$ contour plot

Simulations are to create a map of $C_{\text{SUPP}}$ v $E_{\text{APPLIED}}$ to figure out a better set of paramters that might give full fill inside the feature.

```python
from tools import batch_launch

!date
    
reason = "Running simulation for bulkSuppressor v appliedPotential contour plot with Nx=100"
    
for bulkSuppressor in (2 * np.logspace(-3, -1, 20)):
    for appliedPotential in (-2.5 * np.logspace(-2, 0, 20)):
        batch_launch(reason=reason,
                     tags=['annular54'],
                     kPlus=1e3,
                     kMinus=4e7,
                     rboundary=50e-6 / np.sqrt(np.pi),
                     Nx=100,
                     spacing_ratio=1.2,
                     CFL=0.1,
                     levelset_update_ncell=5,
                     data_frequency=100,
                     bulkSuppressor=bulkSuppressor,
                     appliedPotential=appliedPotential)
```

```python
from multiViewer import MultiViewer
from tools import getRecord

bulkSuppressors = 2 * np.logspace(-3, -1, 20)
appliedPotentials = -2.5 * np.logspace(-2, 0, 20)

reorderedRecords = [[getRecord(records=annular54Records, bulkSuppressor=bulkSuppressor, appliedPotential=appliedPotential) \
    for bulkSuppressor in bulkSuppressors] for appliedPotential in appliedPotentials]

def nonetitle(record, rep, parameter):
    if record:
        return r'{0}={1:1.2e}'.format(rep, record.parameters[parameter])
    else:
        return r'{0}={1}'.format(rep, 'NaN')

viewer = MultiViewer(reorderedRecords,
                     rowtitle=lambda r: nonetitle(r, r'$E$', 'appliedPotential'),
                     columntitle=lambda r: nonetitle(r, r'$c$', 'bulkSuppressor'),
                     figsize=(1.5, 6))

viewer.plot(times=np.arange(20) * 5000. / 19.)

smt_ipy_table([j for i in reorderedRecords for j in i if j],
              fields=['label', 'timestamp', 'parameters', 'duration', 'version', 'tags'],
              parameters=['bulkSuppressor', 'appliedPotential', 'kPlus', 'kMinus'])
```
