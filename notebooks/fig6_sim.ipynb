# Refinement 1

```python
%load_ext autoreload
%autoreload 2
from tools import getSMTRecords, smt_ipy_table
records = getSMTRecords(tags=['annular58'])
```

```python
%load_ext autoreload
%autoreload 4
from featureProperty import FeatureProperty

void_size = []
height = []
kMinus = []
kPlus = []

for r in records:
    if len(r.output_data) > 0:
        fp = FeatureProperty(r)
        void_size.append(fp.getVoidSize())
        height.append(fp.getHeight())
        fp.close()
        kMinus.append(r.parameters['kMinus'])
        kPlus.append(r.parameters['kPlus'])
    
void_size = np.array(void_size)
height = np.array(height)
kPlus = np.array(kPlus)
kMinus = np.array(kMinus)

values = height - void_size
```

```python
from tools import refine_contour_plot
points = np.log10(np.array((kMinus, kPlus)).transpose())
new_points = refine_contour_plot(points, values)[:100]
new_kMinus, new_kPlus = 10**new_points[:,0], 10**new_points[:,1]
```

```python
from IPython.display import Image
Image(filename='fig6_nx200.png')
```

```python
from scipy.spatial import Delaunay
ps = np.concatenate((points, new_points))
tri1 = Delaunay(points)
tri2 = Delaunay(ps)
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax1.triplot(points[:,0], points[:,1], tri1.vertices.copy())
ax1.plot(points[:,0], points[:,1], 'o', ms=3)
ax2 = fig.add_subplot(1, 2, 2)
ax2.triplot(ps[:,0], ps[:,1], tri2.vertices.copy())
ax2.plot(ps[:,0], ps[:,1], 'o', ms=3)
plt.show()
```

```python
from tools import batch_launch

!date
rboundary = 50e-6 / np.sqrt(np.pi)
    
reason = "Running simulation for k+ v k- with Nx=200: refinement 1"
    
for kMinus, kPlus in zip(new_kMinus, new_kPlus):
    batch_launch(reason=reason,
                 tags=['annular58_refinement1'],
                 kPlus=kPlus,
                 kMinus=kMinus,
                 rboundary=50e-6 / np.sqrt(np.pi),
                 Nx=200,
                 spacing_ratio=1.2,
                 CFL=0.1,
                 levelset_update_ncell=5,
                 data_frequency=100,
                 bulkSuppressor=0.06,
                 appliedPotential=-0.45)
```

# Feature Array

```python
%load_ext autoreload
%autoreload 2
from tools import getSMTRecords, smt_ipy_table
records = getSMTRecords(tags=['annular58'])
```

```python
from multiViewer import MultiViewer
from tools import getRecord

kPluses = np.logspace(0, 4, 20)
kMinuses = np.logspace(6, 9, 20)

reorderedRecords = [[getRecord(records=records, kPlus=kPlus, kMinus=kMinus) for kPlus in kPluses] for kMinus in kMinuses]

def me(n):
    s = '%1.1e' % n
    m, e = s.split('e')
    return float(m), int(e)

def rowlabel(r):
    if r:
        m, e = me(r.parameters['kMinus'])
        return '${0:1.1f}\\times 10^{{{1}}}$'.format(m, e)
    else:
        return ''

def columnlabel(r):
    if r:
        m, e = me(r.parameters['kPlus'])
        return '${0:1.1f}\\times 10^{{{1}}}$'.format(m, e)
    else:
        return ''
    
viewer = MultiViewer(reorderedRecords,
                     rowtitle=rowlabel,
                     columntitle=columnlabel,
                     figsize=(1.15, 5.))

viewer.plot(times=np.arange(40) * 5000. / 39.)

smt_ipy_table([j for i in reorderedRecords for j in i if j],
               fields=['label', 'timestamp', 'parameters', 'duration', 'version', 'tags'],
               parameters=['kPlus', 'kMinus', 'bulkSuppressor', 'appliedPotential'])
```
